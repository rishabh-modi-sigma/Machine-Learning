```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
lris_df = pd.read_csv("Iris.csv")
```

```python
lris_df.head()
# lris_df.columns
# lris_df.info()
```

```python
# convert 3 species into 0,1,2
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
lris_df['Species'] = le.fit_transform(lris_df['Species'])
```

```python
lris_df.sample(10)
```

```python
x = lris_df.drop("Species", axis=1)
y = lris_df["Species"]
```

```python
y
```

```python
# Train test split
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.5,random_state=42
)
```

```python
#LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train,y_test)
```

```python
#Knn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
#Naivebayes

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
models = {'LogisticRegression': lr , 'Knn': knn , 'Navie_bayes':nb}

for name,model in models.items():
    y_pred = model.predict(x)
    print(f"\n{name}")
    print("Accuracy_score :",accuracy_score(y,y_pred))
    print("Confusion_matrix :",confusion_matrix(y,y_pred))
    print("classification_report :",classification_report(y,y_pred))
    
```
