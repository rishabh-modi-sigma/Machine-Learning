```python
# ===================== IMPORTS =====================
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
import pandas as pd


# ===================== DATA SPLIT =====================
# X = features, y = target
# (Assuming X and y already defined)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)


# ===================== SCALING =====================
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)   # fit + transform
x_test_scaled = scaler.transform(x_test)          # only transform


# ===================== MODEL =====================
knn = KNeighborsClassifier()


# ===================== HYPERPARAMETER GRID =====================
param_grid = {
    "n_neighbors": [3, 5, 7, 9]
}


# ===================== GRID SEARCH + CV =====================
classifier = GridSearchCV(
    knn,
    param_grid,
    cv=5,
    scoring="recall"   # model selection based on recall
)


# ===================== TRAINING =====================
classifier.fit(x_train_scaled, y_train)


# ===================== PREDICTION =====================
y_pred = classifier.predict(x_test_scaled)


# ===================== METRICS =====================
print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred))
print("Recall :", recall_score(y_test, y_pred))


# ===================== CV RESULTS =====================
res = pd.DataFrame(classifier.cv_results_)
print(res[["param_n_neighbors", "mean_test_score"]])

print("res :",res)
# ===================== BEST PARAMETER =====================
print("Best K value :", classifier.best_params_)

```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Train test split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

# Hyperparameter grid
param_grid = {
    "knn__n_neighbors": [3, 5, 7, 9]
}

# GridSearchCV
classifier = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="recall"
)

# Train
classifier.fit(x_train, y_train)

# Predict
y_pred = classifier.predict(x_test)

# Metrics
print("Accuracy :", accuracy_score(y_test, y_pred) * 100, "%")
print("Precision :", precision_score(y_test, y_pred) * 100, "%")
print("Recall :", recall_score(y_test, y_pred) * 100, "%")

# Best K
print("Best Parameters :", classifier.best_params_)
```
